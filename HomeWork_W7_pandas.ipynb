### Step 1. Import the necessary libraries

```python
import pandas as pd
import sys
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/Euro_2012_stats_TEAM.csv).

```python
BASE_PATH = r'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/Euro_2012_stats_TEAM.csv'
```

### Step 3. Assign it to a variable called euro12.

```python
euro12 = pd.read_csv(BASE_PATH)
```

### Step 5. How many team participated in the Euro2012?

```python
teams = euro12['Team']
```

```python
no_of_teams = len(teams.unique())
msg = f'There are {no_of_teams} teams participated in the Euro2012'

sys.stdout.write(msg)
```

### Step 6. What is the number of columns in the dataset?

```python
cols = euro12.shape[1]
msg = f'The Euro12 DataFrame has {cols} columns'

sys.stdout.write(msg)
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
# see the columns name
euro12.columns
```

```python
discipline_cols = ['Team', 'Yellow Cards', 'Red Cards']
discipline = euro12[discipline_cols]
```

```python
# display the first five rows of discipline
discipline.head()
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'])
```

```python
discipline_sorted
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
yellow_cards_mean = discipline['Yellow Cards'].mean()
msg = f'The mean yellow cards given per team is {round(yellow_cards_mean, 2)} Card/Team'

sys.stdout.write(msg)
```
