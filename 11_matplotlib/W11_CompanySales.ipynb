# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
LINK = r'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'

sales = pd.read_csv(LINK)
sales.head()
```

```python
sales.shape
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
plt.subplots(figsize=(8, 5))
plt.plot(sales['month_number'], sales['total_profit'], linewidth=2)
plt.title('Company profit per month')
plt.xticks(range(1, 13))
plt.xlabel('Month number')
plt.ylabel('Total Profit')
plt.yticks(range(int(1e5), int(5e5 + 1), int(1e5)))

plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
plt.subplots(figsize=(8, 5))
plt.plot(sales['month_number'], sales['total_profit'], linewidth=2, linestyle='--', color='red', marker='o', mfc='black', mec='red', label='Profit data of last year')
plt.title('Company profit per month')
plt.xticks(range(1, 13))
plt.xlabel('Month number')
plt.ylabel('Total Profit')
# plt.ylim(100000, 500000)
plt.yticks(range(int(1e5), int(5e5 + 1), int(1e5)))

plt.legend(loc='lower right')
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
data = sales.iloc[:, :7]
data.head()
```

```python
data.set_index('month_number', inplace=True)
```

```python
yticks = [1000] + [x for x in range(2000, 12001, 2000)] + [15000, 18000]
yticks
```

```python
data.plot(marker='o', linewidth=3, figsize=(10, 6), ms=7)
plt.legend(['Face cream Sales Data', 'Face wash Sales Data', 'Tooth paste Sales Data', 'Bathing Soap Sales Data', 'Shampoo Sales Data', 'Moisturizer Sales Data'])
plt.xticks(range(1, 13))
plt.yticks(yticks)
plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python

tooth_paste = sales[['month_number', 'toothpaste']]
tooth_paste
```

```python
tooth_paste.plot(kind='scatter', x='month_number', y='toothpaste', figsize=(7, 5), s=50)
plt.grid(True, linestyle='--', which='both', linewidth=1.25)
plt.xticks(range(1, 13))
plt.legend(['Tooth paste Data Sales'])
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
plt.subplots(figsize=(10, 6))
plt.bar(sales['month_number'], sales['facecream'], width=0.3)
plt.bar(sales['month_number']+0.3, sales['facewash'], width=0.3)
plt.grid(linestyle='--', which='both', linewidth=1.25)
plt.xticks(sales['month_number'])
plt.gca().set(xlabel='Month Number', ylabel='Sales units in number', title='Facewash and facecream sales data')
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
sales
```

```python
profit = sales['total_profit'].tolist()
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profit, profit_range, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
fig, ax = plt.subplots(figsize=(5, 5))
labels = ['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesList = [sales[x].sum() for x in sales.iloc[:, 1:7]]
ax.pie(salesList, labels=labels, autopct='%1.1f%%')
ax.set_title('Sales data')
plt.legend(loc='lower right', bbox_to_anchor=(1.15, 0.08))
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
face_bathing = sales[['month_number', 'bathingsoap', 'facewash']]
face_bathing.head()
```

```python
face_bathing.set_index('month_number', inplace=True)
face_bathing
```

```python
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(10, 6))
axs[0].plot(face_bathing.index, face_bathing['bathingsoap'], marker='o', color='black', mfc='black', mec='black', linewidth=4, ms=10)
axs[0].set_title('Sales data of a Bathingsoap')
axs[0].set_yticks([7500, 10000, 12500])
axs[1].plot(face_bathing.index, face_bathing['facewash'], marker='o', color='red', mfc='red', mec='red', linewidth=4, ms=10)
axs[1].set_title('Sales data of a facewash')
axs[1].set_ylabel('Sales units in number')
axs[1].set_yticks([1500, 2000])
axs[1].set_xticks(range(1, 13))
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/10.png?raw=1)

```python
data_sales = sales.iloc[:, :7]
data_sales.set_index('month_number', inplace=True)
data_sales.head()
```

```python
data_sales.columns = [
    'face Cream',
    'Face wash', 
    'Tooth paste',
    'Bathing soap',
    'Shampoo',
    'Moisturizer'
]

data_sales.plot.area(color={
    'face Cream': 'purple',
    'Face wash': '#0FFFFF',
    'Tooth paste': 'red',
    'Bathing soap': 'black',
    'Shampoo': 'green',
    'Moisturizer': '#9B870C'
    })
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')

plt.show()
```
