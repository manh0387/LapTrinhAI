### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
LINK = r'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv'

tips = pd.read_csv(LINK)

tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
```

```python
tips
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Distribution of total_bill column')
plt.show()
# Dữ liệu phân bố theo dạng Gaussian distribution (normal distribution) tuy nhiên đang skewed về phía bên trái
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('The scatter plot show the relation ship of total_bill and tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', size='size', hue='size')
plt.title('The relationship of total_bill, tip and size')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.scatterplot(data=tips, x='day', y='total_bill')
plt.title('The relationship between day and total_bill')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(data=tips, x='tip', y='day', hue='sex')
plt.title('Day as the y-axis, tip as the x-axis with sex as hue')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(data=tips, x='total_bill', y='time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.histplot, "tip", kde=True)
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()
```
