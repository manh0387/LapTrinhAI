### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Create the DataFrame that should look like the one below.

```python
first_name = ['Jason', 'Molly', 'Tina', 'Jake', 'Amy']
last_name = ['Miler', 'Jacobson', 'Ali', 'Milner', 'Cooze']
age = [42, 52, 36, 24, 73]
female = [0, 1, 1, 0, 1]
preTestScore = [4, 24, 31, 2, 3]
postTestScore = [25, 94, 57, 62, 70]

mapper = {
    'first_name': first_name,
    'last_name': last_name,
    'age': age,
    'female': female,
    'preTestScore': preTestScore,
    'postTestScore': postTestScore
}
df = pd.DataFrame(mapper)

df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
sns.scatterplot(x = df['preTestScore'], y=df['postTestScore'], size=df['age'])
plt.title('preTestScore and postTestScore\nwith the size of each point determined by age')
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python

sns.scatterplot(x = df['preTestScore'], y=df['postTestScore'], hue=df['female'].apply(lambda x: 'female' if x else 'male'), s=df['postTestScore']*4.5)
plt.title('preTestScore and postTestScore\nwith the size of each point determined by age')
plt.legend(loc='best', bbox_to_anchor=(1.25, 1))
plt.show()
```

### BONUS: Create your own question and answer it.

```python
# Tính tổng điểm trung bình của preTestScore và postTestScore theo nam và nữ
group = df.groupby('female')[['preTestScore', 'postTestScore']].mean()
group
```

```python
group.index = ['male', 'female']
group.plot(kind='bar', rot=0)
plt.xlabel('')
plt.title('Average PreTestScore and PostTestScore')
plt.show()
# Từ biểu đồ cột có thể thấy rằng điểm trung bình của preTestScore và postTestScore của nữ cao hơn của nam
```
